```python
only_delayed_data = only_delayed_data.drop(['Arrival_Delay', 'Delayed','Arr_Temp_3', 'Arr_Pressure_3', 'Arr_Humidity_3',
       'Arr_Wind_Speed_3', 'Arr_Wind_Deg_3', 'Arr_Clouds_3', 'Arr_Rain_3'], axis=1)
```

```python
weight = count
sum = count.sum()
for i in range(len(unique)):
  weight[i] = int(float(count[i])/ float(sum) * 500.0 + 1)
weight 
```

```python
unique
```

```python
only_delayed_data.drop(only_delayed_data[only_delayed_data['Carrier_Code'] == 1].index, inplace = True)
only_delayed_data.drop(only_delayed_data[only_delayed_data['Carrier_Code'] == 2].index, inplace = True)
only_delayed_data.drop(only_delayed_data[only_delayed_data['Carrier_Code'] == 4].index, inplace = True)
only_delayed_data.drop(only_delayed_data[only_delayed_data['Carrier_Code'] == 5].index, inplace = True)
only_delayed_data.drop(only_delayed_data[only_delayed_data['Carrier_Code'] == 7].index, inplace = True)
```

```python
unique, count = np.unique(only_delayed_data.Carrier_Code, return_counts=True)
weight = count
sum = count.sum()
for i in range(len(unique)):
  weight[i] = int(float(count[i])/ float(sum) * 500.0 + 1)
weight 
```

```python
unique
```

```python
stratified_data = pd.DataFrame(columns=['Carrier_Code', 'Origin_Airport', 'Destination_Airport',
       'Flight_Season', 'Flight_Month', 'Flight_Week', 'Flight_Day',
       'Scheduled_Departure_Time', 'Actual_Departure_Time', 'Departure_Delay', 'Dep_Temp_3', 'Dep_Pressure_3',
       'Dep_Humidity_3', 'Dep_Wind_Speed_3', 'Dep_Wind_Deg_3', 'Dep_Clouds_3',
       'Dep_Rain_3'])
for i in range(len(weight)):
  current = only_delayed_data[only_delayed_data.Carrier_Code ==unique[i]]
  current = current.sample(n=weight[i],random_state=1)
  stratified_data = pd.concat([stratified_data, current], ignore_index=True)

stratified_data.sample(5)
```

```python
len(stratified_data)
```

```python
stratified_data = stratified_data.drop(["Destination_Airport"], axis=1)
```

```python
depart_data = pd.read_csv('./depart_combined_2021.csv')
depart_data.sample(5)
```

```python
depart_data = depart_data[depart_data['Actual_Elapsed_Time'] != 0]
```

```python
# airport integerize
airports_code = pd.read_csv('./airports.csv')
depart_destination_counts = depart_data['Destination_Airport'].value_counts().reset_index()
depart_destination_counts.columns = ['Airport', 'Code']
num = depart_destination_counts.shape[0]
depart_destination_counts['Code'] = [num+1]*num - depart_destination_counts.index
airports_code = pd.merge(airports_code, depart_destination_counts, how='outer')
airports_code.drop(['Code (int)'], axis=1, inplace=True)
airports_code.fillna(0, inplace=True)
airports_code
```

```python
airports_code.to_csv("airports_code_with_value.csv", index=False)
```

```python
air_dict = airports_code.set_index('Code').T.to_dict('list')
airport_abbr = []
#airport_loc = []
for index, row in stratified_data.iterrows():
    # print(row)
    # break
    airport_abbr.append(air_dict[row["Origin_Airport"]][0])
airport_abbr
```

```python
stratified_data["Airport_Abbr"] = airport_abbr
```

```python
geo_data = pd.read_csv("./airports_geographic_coordinates.csv")
geo_data.sample(5)
```

```python
geo_dict = geo_data.set_index('Airport').T.to_dict("list")
Lat_N = []
Lng_W = []
GMT = []
for index,row in stratified_data.iterrows():
    Lat_N.append(geo_dict[row["Airport_Abbr"]][0])
    Lng_W.append(geo_dict[row["Airport_Abbr"]][1])
    GMT.append(geo_dict[row["Airport_Abbr"]][2])

stratified_data["Lat_N"] = Lat_N
stratified_data["Lng_W"] = Lng_W
stratified_data["GMT"] = GMT

stratified_data.iloc[0]
```

```python
airlines_dict = {1:'G4' , 2:'F9', 3: 'NK', 4:'HA', 5:'B6', 6:'AA' ,7:'AS', 8:'UA', 9:'WN', 10:'DL'}

airline_abbr = []
for index, row in stratified_data.iterrows():
    airline_abbr.append(airlines_dict[row["Carrier_Code"]])
airline_abbr
```

```python
stratified_data["Carrier_Abbr"] = airline_abbr
```
